*Custom and Pre-trained VGG16 & MLP Models for Image Classification**

##1. Introduction

The project explore the image classification using three neural network architectures: a pre-trained VGG16 model, a custom VGG16 model, and a custom Multi-Layer Perceptron (MLP). Each model is designed to address a classification task involving a dataset of images. The pre-trained VGG16 leverages transfer learning, while  custom VGG16 and the MLP provide insights into the performance of tailored architectures. Hyperparameters such as learning rate, batch size, and hidden layer sizes are adjusted to optimize model performance.

##2. Model Architectures

**Pre-trained VGG16:**

This model has 16 layers, including 13 convolutional layer with the 3x3 filters, 5 max pooling layers, and 3 fully connected layers. It is pre-trained on ImageNet and fine-tuned for the specific classification task.

**Custom VGG16:**

 Similar to pre-trained version, but tailored to the dataset. It includes customized convolutional blocks and adjusted fully connected layers to match  number of output classes.

**Custom MLP:**

A fully connected network with an input layer that flattens 64x64 images, two hidden layers with ReLU activation (the first with 128 neurons), and an output layer that produces probabilities for 8 classes.

## 3. Design Choices

**VGG16:**

VGG16 was chosen for its proven performance in the image classification tasks and its deep architecture, which captures complex features effectively. The option to fine-tune a pre-trained model allows for faster convergence and better generalization, especially with limited training data.

**MLP:**

MLP was selected for its simplicity and efficiency in handling lower-resolution images (64x64 pixels). It serves as a baseline model to compare against VGG16's performance.

**Benefits of Pre-trained Models:**

Using pre-trained models like VGG16 benefits from learned features on large datasets, reducing training time and resource requirements.


## 4. Hyperparameters
Learning Rate: Set to 0.001 to balance convergence speed and stability, based on common practices and initial tests.

**Batch Size:**

Chosen as 16 to optimize memory usage and training stability, allowing effective weight updates without excessive noise.

**Number of Epochs:**

 Set to 10 after monitoring the training and validation loss, ensuring adequate learning without overfitting.

**Hidden Layer Size:**

 For  MLP model, 128 neurons were selected through experimentation, providing effective feature learning while maintaining efficiency.
